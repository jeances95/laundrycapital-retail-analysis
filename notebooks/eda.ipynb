# 📊 Exploratory Data Analysis (EDA)

This analysis explores the Laundry Capital retail dataset to assess data quality, understand distributions, and identify key business insights.

```python
import pandas as pd

df = pd.read_excel("../data/raw/sales_laundrycapital.xlsx")

df.head()
```

```python
df.info()
df.describe()
```

## 🧠 Observations
- Dataset contains transactional sales data
- Numeric fields include revenue and weight (lbs)
- Data types are mostly consistent

```python
df.isnull().sum()
```

## 🔍 Missing Values
- No significant missing values detected
- Data quality is acceptable for analysis

```python
df.duplicated().sum()
```

## 🔁 Duplicates
- Duplicate records identified (if any)
- Handled during ETL process using drop_duplicates()

```python
import matplotlib.pyplot as plt

df['revenue'].hist()
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

## 📊 Revenue Distribution
- Distribution is right-skewed
- Indicates presence of high-value transactions

```python
df.groupby('service_id')['revenue'].sum().sort_values(ascending=False)
```

## 💰 Revenue by Service
- Certain services contribute more significantly to total revenue
- Useful for identifying high-performing service lines

## 📌 EDA Summary

- Dataset contains structured transactional retail data
- No major data quality issues detected
- Revenue distribution shows variability with high-value transactions
- Insights from EDA informed ETL cleaning steps
